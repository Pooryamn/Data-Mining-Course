# Data visulization:

```python
# import libraries: 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import data: 
Amtrak_df = pd.read_csv('../ClassData/Amtrak.csv',squeeze=True)
```

```python
# get some detail from overall data
print('Data schematic :')
print(Amtrak_df.head(9))

print('\nColumns name :')
print(Amtrak_df.columns)

print('\nSize :')
print(Amtrak_df.size)

print('\nShape :')
print(Amtrak_df.shape)
```

```python
# preprocessing data:
#create a new column for date in a suitable format:
Amtrak_df['Date'] = pd.to_datetime(Amtrak_df.Month,format = '%d/%m/%Y')
```

# Plotting Data

```python
# Making Data to plot
RiderShip_ts = pd.Series(Amtrak_df.Ridership.values,index=Amtrak_df.Date)
```

```python
# ploting:
RiderShip_ts.plot(ylim=[1300,2300],legend=False,figsize=[10,5])
plt.xlabel('Year')
plt.ylabel('Rideship (in 1000s)')

plt.tight_layout()
plt.show()
```
